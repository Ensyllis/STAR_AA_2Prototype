```python
import os

# Define the paths to your train and test directories
train_path = r'C:\Users\josep\Coding\Kliest-Identifying-\STAR-Embedding\seventh_attempt\dataset\train'
test_path = r'C:\Users\josep\Coding\Kliest-Identifying-\STAR-Embedding\seventh_attempt\dataset\test'

# Check if train and test directories exist
if os.path.exists(train_path):
    print(f"Train directory found: {train_path}")
else:
    print(f"Train directory not found: {train_path}")

if os.path.exists(test_path):
    print(f"Test directory found: {test_path}")
else:
    print(f"Test directory not found: {test_path}")

# List the text files inside each subfolder for Rowling and Tolkien
for subdir in ['rowling', 'tolkien']:
    subdir_path = os.path.join(train_path, subdir)
    if os.path.exists(subdir_path):
        print(f"\nListing files in {subdir_path}:")
        for filename in os.listdir(subdir_path):
            if filename.endswith(".txt"):
                print(f"- {filename}")
    else:
        print(f"{subdir_path} does not exist.")
```
